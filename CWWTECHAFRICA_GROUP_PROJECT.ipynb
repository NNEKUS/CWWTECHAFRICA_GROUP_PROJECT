```python
import random

def ask_question(question):
    """Ask a single question and return if the answer was correct."""
    print(question["question"])
    for i, option in enumerate(question["options"], 1):
        print(f"{i}. {option}")

    try:
        answer = int(input("Your answer: "))
        if question["options"][answer - 1] == question["answer"]:
            print("Correct!\n")
            return True
        else:
            print(f"Wrong! The correct answer is: {question['answer']}\n")
            return False
    except (IndexError, ValueError):
        print(f"Invalid input! The correct answer is: {question['answer']}\n")
        return False

def run_quiz(questions):
    """Run the quiz and return the user's score."""
    score = 0
    for question in questions:
        if ask_question(question):
            score += 1
    return score

def calculate_result(score, total_questions, pass_mark):
    """Calculate and print the quiz result."""
    percentage = (score / total_questions) * 100
    print(f"\nYour score: {score}/{total_questions}")
    print(f"Percentage: {percentage:.2f}%")
    if percentage >= pass_mark:
        print("Congratulations! You passed the quiz.")
    else:
        print("You failed the quiz. Better luck next time!")

def main():
    """Main function to run the quiz."""
    questions = [
        {
            "question": "What is the capital of France?",
            "options": ["Berlin", "Madrid", "Paris", "Lisbon"],
            "answer": "Paris"
        },
        {
            "question": "Who wrote 'To Kill a Mockingbird'?",
            "options": ["Harper Lee", "Mark Twain", "Jane Austen", "Charles Dickens"],
            "answer": "Harper Lee"
        },
        {
            "question": "What is the chemical symbol for water?",
            "options": ["O2", "H2O", "CO2", "NaCl"],
            "answer": "H2O"
        },
        {
            "question": "Which planet is known as the Red Planet?",
            "options": ["Earth", "Venus", "Mars", "Jupiter"],
            "answer": "Mars"
        },
        {
            "question": "What is the largest ocean on Earth?",
            "options": ["Atlantic Ocean", "Indian Ocean", "Arctic Ocean", "Pacific Ocean"],
            "answer": "Pacific Ocean"
        }
    ]

    random.shuffle(questions)  # Optional: Shuffle questions to make the quiz different each time

    total_questions = len(questions)
    pass_mark = 60.0  # Pass mark in percentage

    score = run_quiz(questions)
    calculate_result(score, total_questions, pass_mark)

if __name__ == "__main__":
    main()
```
